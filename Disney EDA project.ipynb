## 4.Data Analysis & Visualization 

### i.Content distribution over time

```python
sns.histplot(x=data['type'], y=data['release_year'], color='darkblue')
plt.xlabel('Type')
plt.ylabel('Year')
plt.title('Release Years and Types')
plt.show()
```

*Disney released the most number of movies and TV shows in 2020. This suggests that Disney is currently investing heavilyinnewcontent.

*Disney released a relatively equal number of movies and TV shows in 1980 and 2000. This suggests that Disney was focused on producing both movies and TV shows inthesedecades.

*Disney has more movies released in year 1940-1980 compared to TV shows. This suggests that Disney was focused on streaming old movies rather than TV shows inthesedecades.

### ii.Count of different types of shows in Disney

```python
colour=['darkblue','lightblue']
sns.countplot(x='type',data=data,palette=colour)
plt.title('Count of Different shows on Disney')
plt.show()
```

*Disney released the more number of movies Than TV shows. This suggests that Disney is investing heavily in movies.

### iii. Count of movies and TVshows over year

```python
count=data['release_year'].value_counts()
count
```

```python
sns.kdeplot(data['release_year'],shade=True)
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title("Distribution of Release Years")
plt.show()
```

*Disney released more number of movies and TV shows during the period 2000-2020

*Disney have highest number of release in 2021

### iv. Top 10 countries with highest contents

```python
country=data['country'].value_counts().head(10)
country
```

```python
plt.barh(country.index,country.values)
plt.xlabel('Content Count')
plt.ylabel('Country')
plt.title('Content Counts by Country')
plt.gca().invert_yaxis()  # Invert y-axis to have the country with the highest count at the top
plt.show()




```

*Most of the contents in Disney is from United States

### vii.Genres in Disney

```python
#extracting genres
genres = ', '.join(data['listed_in']).split(', ')
unique_genres = list(set(genres))
#counting genres
count_g={i:genres.count(i) for i in unique_genres}
count_g
```

### vii.Top 10 Genres

```python
#extracting top 10 genre
top_genres = dict(sorted(count_g.items(), key=lambda x: x[1], reverse=True)[:10])

# Create a horizontal bar plot
sns.barplot(x=list(top_genres.values()), y=list(top_genres.keys()))
plt.xlabel('Number of Contents')
plt.ylabel('Genre')
plt.title('Top 10 Genres')
plt.show()
```

*Disney have more family and animated contents

### ix.Top 10 Actors with most contents

```python
#extracting 
cast = ', '.join(data['cast']).split(', ')
unique_cast = list(set(cast))
#counting genres
count_c={i:cast.count(i) for i in unique_cast}
top_cast = dict(sorted(count_c.items(), key=lambda x: x[1], reverse=True)[:10])
top_cast
```
